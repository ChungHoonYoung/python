영상 (웹캠) 사용하기

```python
# 웹캠 입력 예제

import cv2

cap = cv2.VideoCapture(0)

while (cap.isOpened()) :

    ret, img = cap.read()

    cv2.imshow('myWin', img)

    if cv2.waitKey(1) & 0xFF == ord('q') :

        break

    print(ret, img)

cap.release()
cv2.destroyAllWindows()
```

```python
# 웹캠 영상 저장

import cv2

cap = cv2.VideoCapture(0)

cap.set(cv2.CAP_PROP_FRAME_WIDTH, 320)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 240)

fourcc = cv2.VideoWriter_fourcc(*'DIVX')
writer = cv2.VideoWriter('video.avi', fourcc, 25.0, (320, 240))

while (cap.isOpened()) :
    ret, img = cap.read()
    cv2.imshow('myWin', img)
    writer.write(img)

    if cv2.waitKey(1) & 0xFF == ord('q') :
        break
    print(ret, img)

cap.release()
cv2.destroyAllWindows()
```

```python
# 저장된 영상 재생하기

import cv2
cap = cv2.VideoCapture('video.avi')

while cap.isOpened() :
    ret, frame = cap.read()
    if not ret :
        print("Is showtime")
        break

    cv2.imshow('video', frame)

    if cv2.waitKey(20) == ord('q') :
        print("Is shutdown")
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# 마우스로 만들기

import numpy as np
import cv2

def draw_circle(event, x, y, flages, param):
    if event == cv2.EVENT_LBUTTONDBLCLK :
        cv2.circle(img, (x, y), 100, (0, 255, 255), -1)
    if event == cv2.EVENT_RBUTTONDBLCLK :
        cv2.circle(img, (x, y), 50, (0, 255, 0), -1)

img = np.zeros((512, 512, 3),dtype="uint8" )
cv2.namedWindow('image')
cv2.setMouseCallback('image', draw_circle)

while(1) :
    cv2.imshow('image', img)
    if cv2.waitKey(20) & 0xFF == 27 :
        break

cv2.destroyAllWindows()
```

```python
# 드래그하면 사각형, m키를 누르면 원과 사각형으로 모드 변경

import numpy as np
import cv2

drawing = False
mode = True
ix, iy = -1, -1

def draw_circle(event, x, y, flages, param):
    
    global ix, iy, drawing, mode

    if event == cv2.EVENT_LBUTTONDOWN :
        drawing == True
        ix, iy = x, y

    elif event == cv2.EVENT_MOUSEMOVE :
        if drawing == True :
            if mode == True :
                cv2.rectangle(img, (ix, iy), (x, y), (255, 0, 0), -1)
            else :
                cv2.circle(img, (ix, iy), 5, (0, 255, 0), -1)

    elif event == cv2.EVENT_LBUTTONUP :
        drawing == False
        if mode == True :
            cv2.rectangle(img, (ix, iy), (x, y), (255, 0, 0), -1)
        else :
            cv2.circle(img, (x, y), 5, (0, 255, 0), -1)

img = np.zeros((512, 512, 3),dtype="uint8" )
cv2.namedWindow('image')
cv2.setMouseCallback('image', draw_circle)

while True :
    cv2.imshow('image', img)
    k = cv2.waitKey(1) & 0xFF

    if k == ord('m') :
        mode = not mode
    elif k == 27 :
        break

cv2.destroyAllWindows()
```

```python
# RGB 값 생성기 / createTrackbar로 슬라이드 만들기

import cv2
import numpy as np

def nothing(x):
    pass

img = np.zeros((300,512,3), np.uint8)
cv2.namedWindow('image')

# trackbar를 생성하여 named window에 등록
cv2.createTrackbar('R', 'image', 0, 255, nothing)
cv2.createTrackbar('G', 'image', 0, 255, nothing)
cv2.createTrackbar('B', 'image', 0, 255, nothing)

switch = '0:OFF\n1:On'
cv2.createTrackbar(switch, 'image', 1, 1, nothing)

while(1):
    cv2.imshow('image', img)

    if cv2.waitKey(1) & 0xFF == 27:
        break

    r = cv2.getTrackbarPos('R','image')
    g = cv2.getTrackbarPos('G', 'image')
    b = cv2.getTrackbarPos('B', 'image')
    s = cv2.getTrackbarPos(switch, 'image')

    if s == 0:
        img[:] = 0 # 모든 행/열 좌표 값을 0으로 변경. 검은색
    else:
        img[:] = [b,g,r] # 모든 행/열 좌표값을 [b,g,r]로 변경

cv2.destroyAllWindows()
```

```python
# 이미지의 연산 : 더하기

import cv2
import numpy as np

img1 = cv2.imread('./1_10______/flower1.jpg')
img2 = cv2.imread('./1_10______/flower2.jpg')

while True :
    img3 = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)
    
    cv2.imshow("Screen", img3)
    
    if cv2.waitKey(1) & 0xFF == 27 :
        break
    
cv2.destroyAllWindows()
```

```python
# 이미지 사이즈 변경

import cv2

img = cv2.imread('./1_10______/flower1.jpg', cv2.IMREAD_COLOR)
print(img.shape)

resized_img_1 = cv2.resize(img, dsize=(200, 200), interpolation=cv2.INTER_LINEAR)
print(resized_img_1.shape)

resized_img_2 = cv2.resize(img, dsize=(500, 500), interpolation=cv2.INTER_LINEAR)
print(resized_img_2.shape)

cv2.imshow("resized_img_1", resized_img_1)
cv2.imshow("resized_img_2", resized_img_2)

key = cv2.waitKey(0)

if key == 27 :  # esc(아스키코드로 27번) 누르면 저장
    cv2.imwrite("resized_img_1.jpg", resized_img_1)
    cv2.imwrite("resized_img_2.jpg", resized_img_2)

cv2.destroyAllWindows()
```
